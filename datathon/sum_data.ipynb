```python
import pandas as pd
import pickle

with open('variables.pkl', 'rb') as f:
    data_AA, data_BB, data_CC, data_DD, data_after_A = pickle.load(f)
#%run /Users/John/Desktop/datathon/Clear_Column.ipynb
%run "Clear_Column.ipynb"
#clean leftover data that is good
#pd.options.display.max_rows = None


print(data_AA,"ewiduwvubcihvweufcggwqbvehcyw")

def clear_data_All(dataframe):
    dataframe = dataframe[["State"]]
    return dataframe

#combine all data
def combine_data(df1sorted, column_a, column_b, column_c, column_d):
    sum_data = pd.concat([df1sorted, column_a, column_b, column_c, column_d], axis = 0)
    return sum_data

#count faciltiies in each state
def count_facilities(sum_data):
    facilities_sum = sum_data.pivot_table(index = ["State"], aggfunc ='size')
    return facilities_sum



rest_data = clear_data_All(data_after_A)
print("rest data", rest_data)

combine = combine_data(rest_data, data_DD, data_CC, data_BB, data_AA)
print("combine data",combine)

combined_facilities = count_facilities(combine)

print(combined_facilities)
```
